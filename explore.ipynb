### Figuring out the sensors


```python
sensor_df = df.filter(regex="sensor")
sensor_df
```

```python
s1 = sensor_df["sensor 1"].apply(pd.Series)
s1
```

```python
s1["value_type"].unique()
```

```python
# for each columns in sensor_df, apply pd.Series and show a few non null values
list_of_dfs = []
for col in sensor_df.columns:
    list_of_dfs.append(sensor_df[col].apply(pd.Series).dropna())
```

```python
# The first df has the sensor name and type
sensor_types = list_of_dfs.pop(0)
```

```python
sensor_types.name.unique()
```

According to my research, the most important sensors are SDS011 and SPS30


```python
sensor_types[sensor_types["name"].isin(["SDS011", "SPS30"])].shape
# Out of 510 rows, 283 are sensors collecting air quality data
```

```python
for df in list_of_dfs:
    # get the unique value types
    print("====================================")
    try:
        value_types = df["value_type"].unique()
        print(
            f"""Value types: {value_types} 
DF shape: {df.shape} """
        )
    except KeyError:
        print("No value_type column")
        print(df)

# The most populated pins are pin 1 and 2
```

It looks like the 2 first sensor columns are the most important ones


```python
def explode_column(df, column):
    exploded = df[column].apply(pd.Series)
    df = pd.concat([df, exploded], axis=1)
    df = df.drop(columns=[column])
    return df


sensor_df = explode_column(sensor_df, "sensor_type")
```

```python
sensor_df.groupby(["name", "manufacturer"]).count()
```

```python
# Check what type of sensors are for the SDS011

for sensor in sensor_df[sensor_df["name"] == "SDS011"][["sensor 0", "sensor 1"]].iloc[
    0
]:
    print(sensor)
```

### Decisions

We're going to drop all the sensors other than SDS011, there's no point in saving the temperature + humidity and the other useful sensor does not appear very often (SPS30)


```python
df = pd.read_json("message.json", lines=True)
df = explode_weather_data(df)
```

```python
df.columns[13:24]
```

```python
# drop all columns from sensor 2 to sensor 12
df = df.drop(columns=df.columns[13:24])
df.columns
```

```python
from helper import explode_column

df = explode_column(df, "sensor_type")
```

```python
# set null for all names other than sds011
df.loc[df["name"] != "SDS011", ["name", "sensor 0", "sensor 1"]] = None
```

```python
df.groupby(["name", "manufacturer"]).count()
```

```python
df["sensor 0"].isna().sum()
```

```python
# Extract value from sensor 0 and 1, rename them to p1 and p2
df["p1"] = df["sensor 0"].apply(lambda x: pd.Series(x).get("value"))
df["p2"] = df["sensor 1"].apply(lambda x: pd.Series(x).get("value"))

df = df.drop(columns=["sensor 0", "sensor 1"])

df
```
